## Introduction to numpy and matplotlib

## Today, we are performing straight line equation "y=mx+c" where 'm' denotes the slope and c is constant.


```python
m=5
c=-1
x=[0,1,2,3,4,5,6]
for x_i in x:
    y = [m*x_i + c]
    print(y)
```

## Indroduce the Numpy and Matplotlib

```python
import numpy as np
```

#numpy syntax as np.array for the list.

```python
X = np.array(x)
Y = m*X + c
```

```python
print(Y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,Y)
```

```python
plt.plot(X,Y)
```

```python
plt.xscale('log')
```

```python
plt.hist(X,bins=5)
```

## X_labbel,Y_labbel and Header in Plot 

```python
plt.scatter(X,Y)
plt.xlabel('X_axix')
plt.ylabel('Y_axix')
plt.title('Straight line Y=m*x+c')
```
